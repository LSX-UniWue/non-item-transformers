## Check Coveo Datasets for Nan values in each column

```python
import pandas as pd
```

---
### Without Pageviews

```python
train_wo_pageviews = pd.read_csv("../coveo_dataset/preprocessed_data/split/without_pageviews/coveo.train.csv", sep="\t")
validation_wo_pageviews = pd.read_csv("../coveo_dataset/preprocessed_data/split/without_pageviews/coveo.validation.csv", sep="\t")
test_wo_pageviews = pd.read_csv("../coveo_dataset/preprocessed_data/split/without_pageviews/coveo.test.csv", sep="\t")
```

```python
coveo_full_wo_pageviews = pd.concat([train_wo_pageviews, validation_wo_pageviews, test_wo_pageviews])
coveo_full_wo_pageviews
```

```python
coveo_full_wo_pageviews.isnull().any()
```
