#### Create Table with all articles in transactions

```python
articles_in_transactions = pd.merge(transactions['article_id'].drop_duplicates(), articles, on='article_id', how='left')
articles_in_transactions
```

---
#### Product Type Stats

```python
product_type_count: pd.DataFrame = articles_in_transactions['product_type_name'].value_counts().reset_index()
product_type_count.columns = ['product_type_name', 'count']

product_type_percentage: pd.DataFrame = articles_in_transactions['product_type_name'].value_counts(normalize=True).reset_index()
product_type_percentage.columns = ['product_type_name', 'percentage']

pd.merge(product_type_count, product_type_percentage, on="product_type_name")
```

#### Product Group Stats

```python
product_group_count: pd.DataFrame = articles_in_transactions['product_group_name'].value_counts().reset_index()
product_group_count.columns = ['product_group_name', 'count']

product_group_percentage: pd.DataFrame = articles_in_transactions['product_group_name'].value_counts(normalize=True).reset_index()
product_group_percentage.columns = ['product_group_name', 'percentage']

pd.merge(product_group_count, product_group_percentage, on="product_group_name")
```

```python
plt.figure(figsize=(25,10))
plt.bar(product_group_count['product_group_name'], product_group_count['count'], width=0.8)
plt.margins(x=0.01, tight=True)
plt.xlabel("Product Group Name", fontsize='x-large')
plt.ylabel("Count", fontsize='x-large')
plt.xticks(rotation=30)
plt.grid()
plt.show()
```

### New articles per day

```python
articles_per_day_list: pd.DataFrame = transactions.groupby('t_dat').agg({'article_id': set}).reset_index()
articles_per_day_list.columns = ['t_dat', 'article_ids']
articles_per_day_list
```

```python
discovered_articles = set()
for day in articles_per_day_list['t_dat']:
    articles_set = articles_per_day_list.loc[articles_per_day_list['t_dat'] == day, 'article_ids'].iloc[0]
    articles_set -= discovered_articles
    discovered_articles.update(articles_set)
articles_per_day_list
```

```python
articles_per_day_list['article_ids'] = articles_per_day_list['article_ids'].apply(lambda x: len(x))
articles_per_day_list
```

Drop first day

```python
articles_per_day_list.drop([0], inplace=True)
```

```python
plt.figure(figsize=(15,8))
plt.plot(articles_per_day_list.index.to_series(), articles_per_day_list['article_ids'])
plt.margins(x=0.01, tight=True)
plt.xlabel("Day")
plt.ylabel("Number of new discovered Articles")
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
