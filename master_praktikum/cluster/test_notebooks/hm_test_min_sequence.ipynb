```python
import pandas as pd
```

```python
train = pd.read_csv("../hm_dataset/preprocessed_data/split/hm.train.csv", sep="\t")
train
```

```python
train.groupby(['customer_id']).size().describe()
```

```python
aggregated = train.groupby(['customer_id']).size()
aggregated
```

```python
filtered = aggregated.apply(lambda v: v >= 2)
filtered = filtered.reset_index()
filtered.columns = ['customer_id', 'min_sequence_bool']
filtered
```

```python
ids = filtered[filtered['min_sequence_bool']]['customer_id'].tolist()
ids
```

```python
train = train[train['customer_id'].isin(ids)].copy()
train
```
