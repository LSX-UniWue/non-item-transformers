# Preprocessing H&M Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
articles = pd.read_csv("../hm_dataset/raw_data/articles.csv")
customers = pd.read_csv("../hm_dataset/raw_data/customers.csv")
transactions = pd.read_csv("../hm_dataset/raw_data/transactions_train.csv")
```

---
#### Convert dates to timestamps

```python
import time
import datetime

def datetime_to_timestamp(date):
    return time.mktime(datetime.datetime.strptime(date, "%Y-%m-%d").timetuple())
    
transactions['t_dat'] = transactions['t_dat'].apply(datetime_to_timestamp)

transactions
```

#### Remove redundant columns in articles

```python
articles.drop(
    inplace=True,
    columns=['product_code', 'product_type_no', 'graphical_appearance_no', 'colour_group_code', 'perceived_colour_value_id', 'perceived_colour_master_id', 'department_no',
            'index_code', 'index_group_no', 'section_no', 'garment_group_no'])
```

## Merge Tables

```python
full_dataset = pd.merge(transactions, customers, on='customer_id', how='left')
full_dataset = pd.merge(full_dataset, articles, on='article_id', how='left')
full_dataset.sort_values(['customer_id', 't_dat'], inplace=True)
full_dataset
```

```python
full_dataset.to_csv(path_or_buf="../hm_dataset/preprocessed_data/hm_preprocessed.csv", sep="\t", index=False)
```
