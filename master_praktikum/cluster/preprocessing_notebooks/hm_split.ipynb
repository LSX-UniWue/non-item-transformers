# H&M Dataset Split

```python
import pandas as pd
```

```python
full_dataset = pd.read_csv("../hm_dataset/preprocessed_data/hm_preprocessed.csv", sep="\t")
```

```python
full_dataset.sort_values(['t_dat'], inplace=True)

full_dataset
```

---
#### Remove nan values

```python
full_dataset.fillna(0, inplace=True)

full_dataset
```

---
#### Apply item feedback

```python
full_dataset.groupby(['article_id']).size().describe()
```

```python
aggregated = full_dataset.groupby(['article_id']).size()
filtered = aggregated.apply(lambda v: v >= 5)
filtered = filtered.reset_index()
filtered.columns = ['article_id', 'item_feedback_bool']
ids = filtered[filtered['item_feedback_bool']]['article_id'].tolist()
full_dataset = full_dataset[full_dataset['article_id'].isin(ids)].copy()
```

```python
full_dataset.groupby(['article_id']).size().describe()
```

---
### Create Splits and save split files

```python
end_of_train = 1581379200
end_of_validation = 1591660800
```

```python
train = full_dataset.loc[(full_dataset['t_dat']<=end_of_train)].copy()
validation = full_dataset.loc[(full_dataset['t_dat']<=end_of_validation) & (full_dataset['t_dat']>end_of_train)].copy()
test = full_dataset.loc[(full_dataset['t_dat']>end_of_validation)].copy()
```

```python
train.sort_values(['customer_id', 't_dat'], inplace=True)
validation.sort_values(['customer_id', 't_dat'], inplace=True)
test.sort_values(['customer_id', 't_dat'], inplace=True)
```
