Product Type Stats

```python
product_type_count: pd.DataFrame = articles['product_type_name'].value_counts().reset_index()
product_type_count.columns = ['product_type_name', 'count']

product_type_percentage: pd.DataFrame = articles['product_type_name'].value_counts(normalize=True).reset_index()
product_type_percentage.columns = ['product_type_name', 'percentage']

pd.merge(product_type_count, product_type_percentage, on="product_type_name")
```

Product Group Stats

```python
product_group_count: pd.DataFrame = articles['product_group_name'].value_counts().reset_index()
product_group_count.columns = ['product_group_name', 'count']

product_group_percentage: pd.DataFrame = articles['product_group_name'].value_counts(normalize=True).reset_index()
product_group_percentage.columns = ['product_group_name', 'percentage']

pd.merge(product_group_count, product_group_percentage, on="product_group_name")
```

```python
plt.figure(figsize=(25,5))
plt.bar(product_group_count['product_group_name'], product_group_count['count'])
plt.xlabel("Product Group Name")
plt.ylabel("Count")
plt.grid()
plt.show()
```

New articles per day

```python
articles_per_day_list: pd.DataFrame = transactions.groupby('t_dat').agg({'article_id': set}).reset_index()
articles_per_day_list.columns = ['t_dat', 'article_ids']
articles_per_day_list
```

```python
discovered_articles = set()
for day in articles_per_day_list['t_dat']:
    articles_set = articles_per_day_list.loc[articles_per_day_list['t_dat'] == day, 'article_ids'].iloc[0]
    articles_set -= discovered_articles
    discovered_articles.update(articles_set)
articles_per_day_list
```

```python
articles_per_day_list['article_ids'] = articles_per_day_list['article_ids'].apply(lambda x: len(x))
articles_per_day_list
```

```python
articles_per_day_list.drop([0], inplace=True)
```

```python
plt.figure(figsize=(10,5))
plt.bar(articles_per_day_list['t_dat'], articles_per_day_list['article_ids'])
plt.xlabel("Day")
plt.ylabel("Number of new discovered Articles")
plt.grid()
plt.show()
```

Session length stats (with percentiles)

```python
transactions.groupby(["customer_id"]).size().describe()
```

```python
session_lengths: pd.DataFrame = transactions.groupby(["customer_id"]).size().reset_index()
session_lengths.columns = ['customer_id', 'session_length']

session_lengths
```

```python
session_length_count: pd.DataFrame = session_lengths.groupby(["session_length"]).size().reset_index()
session_length_count.columns = ['session_length', 'count']

session_length_count
```

```python
plt.figure(figsize=(10,5))
plt.bar(session_length_count['session_length'], session_length_count['count'])
plt.xlabel("Session Length")
plt.ylabel("Number of Sessions")
plt.xticks(range(1, session_length_count['session_length'].max() + 1))
plt.yticks(range(0, session_length_count['count'].max() + 1, 10))
plt.grid(visible=True, which='major', axis='y')
plt.show()
```

```python
articles_frequency: pd.DataFrame = transactions.groupby(["article_id"]).size().reset_index()
articles_frequency.columns = ['article_id', 'frequency']

articles_frequency.head(10)
```
